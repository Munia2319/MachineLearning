```python
# Dataset is now stored in a Pandas Dataframe
import io
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['test_scores.csv']))
df
```

```python
import numpy as np
import math
def gradient_descent(x,y):
    m_current=b_current=0
    iteration=124400
    learning_rate=0.0002
    n=len(x)
    cost=[]
    y_predicted=m_current*x+b_current
    cost.append((1/n)*sum([val**2 for val in (y-y_predicted)]))
    md=-(2/n)*sum(x*(y-y_predicted))
    bd=-(2/n)*sum(y-y_predicted)
    m_current=m_current-learning_rate*md
    b_current=b_current-learning_rate*bd
    print("m {},b {},iteration 0 cost {}".format(m_current,b_current,cost))
    j=0
    for i in range(iteration):
      j=i+1
      y_predicted=m_current*x+b_current
      cost.append((1/n)*sum([val**2 for val in (y-y_predicted)]))
      md=-(2/n)*sum(x*(y-y_predicted))
      bd=-(2/n)*sum(y-y_predicted)
      m_current=m_current-learning_rate*md
      b_current=b_current-learning_rate*bd
      print("m {},b {},iteration {} cost {}".format(m_current,b_current,j,cost[i]))
      if math.isclose(cost[j], cost[j-1], rel_tol=1e-09, abs_tol=0.0):
        print("Optimum value is gotten")
        print("The final value of m= {} and b= {} and iteration {}".format(m_current,b_current,j))
        break

x=np.array(df.math.tolist())
y=np.array(df.cs.tolist())
y
gradient_descent(x,y)

```
