```python
import pandas as pd
df=pd.read_csv('/content/titanic.csv')
df.head()
```

```python
df[['Sex','Age','Fare','Pclass']].isnull().values.any()
result=df[['Sex','Age','Fare','Pclass']].isnull().sum()
value=df[['Sex','Age','Fare','Pclass']].isnull()
value.head(10)
```

```python
target=df['Survived']
target
```

```python
inputs=df.drop(['Survived','PassengerId','SibSp','Parch','Ticket','Cabin','Embarked','Name'],axis='columns')
inputs
```

```python
#handling with nan values
inputs['Sex'].isnull().values.any()
```

```python
#handling with nan values
inputs['Age'].isnull().values.any()
```

```python
#filling nan value with medians
inputs['Age']=inputs['Age'].fillna(inputs['Age'].mean())
inputs['Age']
```

```python
#checking if still any value is nan
inputs['Age'].isnull().values.any()
```

```python
#now converting string to number
from sklearn import preprocessing
le_sex=preprocessing.LabelEncoder()
```

```python
inputs['Fare'].isnull().values.any()
```

```python
inputs['Pclass'].isnull().values.any()
```

```python
target.isnull().values.any()
import numpy as np
np.isfinite(target.all())
```

```python
#assigning new column to inputs
inputs['sex_n']=le_sex.fit_transform(inputs['Sex'])
inputs.head()
```

```python
inputs_n=inputs.drop('Sex',axis='columns')
inputs_n
```

```python
np.isfinite(inputs_n.all())
np.isnan(inputs_n)
```

```python
#splitting data set for training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs_n,target,test_size=0.2)
```

```python
#creating model
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
