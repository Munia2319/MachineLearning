```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target']=iris.target
df.head(10)
```

```python
iris.target_names
```

```python
#how many rows in target has 1
df[df.target==1].head()
```

```python
#creating new column from existing column in the dataframe
df['flowers_name']=df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
#drawing scatterplot
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='red',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker='.')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='green',marker='+')
```

```python
#drawing scatterplot
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='red',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',marker='.')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='green',marker='+')
```

```python
inputs=df.drop(['target','flowers_name'],axis='columns')
inputs
```

```python
output=df['target']
output
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,output,test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='rbf')
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[3,2,1,2]])
```
